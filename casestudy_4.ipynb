```python
#3.#Check whether all the zones generate the same amount of sales.
```

```python
#One Way Anova Hypothesis
#H0:All zones generate same amount of sales
#H1:All zones generate different amount of sales
#if p-value is less than the (0.05) we will reject the null hypothesis
```

```python
import scipy.stats as stats
fvalue,pvalue = stats.f_oneway(data1['Zone - A'],data1['Zone - B'],data1['Zone - C'],data1['Zone - D'])
print("fvalue=",fvalue, "pvalue=",pvalue)
```

```python
if pvalue<0.05:
   print('Reject the null hypothesis')
else:
   print('Null hypothesis accepted')
```

```python
#Here p-value is less than the (0.05) we will reject the null hypothesis.so 4 zones generate different amount of sale
```
