```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r'C:\Users\jithi\Desktop\Vinu DSA\activity\Telco-Customer-Churn.xlsx - Telco-Customer-Churn.csv')
```

```python
#1. Compare churn count with respect to gender.

sns.countplot(x='gender', hue='Churn', data=dataset, palette=['#FF0000',"#000000"])
```

```python
#2. Find out how many female senior citizens there in the dataset

conditions=[(dataset['SeniorCitizen']==1) & (dataset['gender']=='Female')]
values=['Female Senior citizen']

dataset['count_female_senior_citizen']=np.select(conditions, values)
abc = dataset['count_female_senior_citizen'].value_counts()

dataset['Category_seniorcitizen'].value_counts(normalize=True).plot.pie(autopct=lambda x: '{:.0f}'.format(x*abc.sum()/100) , ylabel='', legend=False, title='Number of Female Senior Citizen')
```

```python
#3. Compare 'tenure' with 'Total Charges'

sns.lineplot(data=dataset, x="tenure", y="TotalCharges", color='blue')
```

```python
4. Find out which contract is preferred by the senior citizen.

dataset.loc[dataset['SeniorCitizen']==0] 
sns.catplot(x='Contract', kind="count", data=dataset).set(title='Contract preferred by the senior citizen')
```
